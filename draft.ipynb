```python
import pandas as pd
import os
from string import printable
from collections import Counter
import numpy as np
```

```python
title=['Interpreting Sustainability','Sustainability reporting']
```

```python

printable_dict = {c:i for i,c in enumerate(printable)}

def vectorize(title):
    """ Returns a vector with counts for each printable character found in a title """
    cnt = Counter(title)
    array = [0 for _ in printable]+[0] # Last position is for nonprintable characters
    for c,n in cnt.items():
        if c in printable_dict:
            array[printable_dict[c]]=n
        else:
            array[-1]+=n # Update last position if it's not printable
    return np.array(array)



vects = []
for title in title:
    vects.append(vectorize(title))
```

```python

from string import printable
from collections import Counter
import numpy as np

printable_dict = {c:i for i,c in enumerate(printable)}

def vectorize(title):
    """ Returns a vector with counts for each printable character found in a title """
    cnt = Counter(title)
    array = [0 for _ in printable]+[0] # Last position is for nonprintable characters
    for c,n in cnt.items():
        if c in printable_dict:
            array[printable_dict[c]]=n
        else:
            array[-1]+=n # Update last position if it's not printable
    return np.array(array)

vects = []
for title in df.title:
    vects.append(vectorize(title))
```

```python
def compute_sum_numpy(array):
  order = np.arange(1, len(array) + 1)
  # Calculate element-wise product of the array and order
  result = np.multiply(array, order)
  return np.sum(result)

```

```python
sum_per_array = np.apply_along_axis(compute_sum_numpy, 1, vects)
print(sum_per_array)
```
