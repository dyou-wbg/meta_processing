#  3.Concate all colnames

```python

def create_dataframe_of_column_names(directory):
  #files = os.listdir(directory)
  files = files_clean
  dataframes = []

  for file in files:
    if file.endswith('.csv') or filename.endswith('.tsv'):
      df = pd.read_csv(os.path.join(directory, file), nrows=5)
      column_names = df.columns

      df_column_names = pd.DataFrame(column_names)
      df_column_names['filename'] = file
      dataframes.append(df_column_names)

  df_all_column_names = pd.concat(dataframes)
  return df_all_column_names

directory = '/content/drive/MyDrive/finalMetaFile'
df_all_column_names = create_dataframe_of_column_names(directory)
print(df_all_column_names)
```
